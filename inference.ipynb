```python
import pandas as pd

hdcn = pd.read_csv("./output/hdcn.csv")
hybrid_gdcn = pd.read_csv("./output/hybrid_gdcn.csv")
xgb = pd.read_csv("./output/xgb_infer_dev.csv")

hdcn_pb = hdcn['clicked']
hybrid_gdcn_pb = hybrid_gdcn['clicked']
xgb_pb = xgb['clicked']
```

-  만약 아래 점수가 오류가 난다면, clicked > 1 이 존재한다면, 그아래 셀로 부탁드립니다.

```python
import pandas as pd
preds = (hdcn_pb*0.5 + hybrid_gdcn_pb*0.5)*0.815 + xgb['clicked'] *0.205
samp = pd.read_csv("sample_submission.csv")
samp['clicked'] = preds
samp.to_csv("./output/check.csv", index=False)
```
