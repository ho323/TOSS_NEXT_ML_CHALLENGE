```python
import numpy as np
import pandas as pd
import xgboost as xgb
import joblib

# =========================================================
# 데이터 로드 및 전처리
# =========================================================
test = pd.read_csv("./test_.csv")
test['gender'].fillna(2, inplace=True)
test['age_group'].fillna(1, inplace=True)
test.drop(columns=["l_feat_17", "seq"], inplace=True)
test.fillna(0, inplace=True)

cat_cols = ['gender', 'age_group', 'inventory_id', 'day_of_week', 'hour']
test[cat_cols] = test[cat_cols].fillna("UNK")
test = pd.get_dummies(test, columns=cat_cols)

# =========================================================
# 모델 및 피처 불러오기
# =========================================================
feature_cols = joblib.load("xgb_feature_cols.pkl")
for c in feature_cols:
    if c not in test.columns:
        test[c] = 0
test = test[feature_cols]

# =========================================================
# 예측
# =========================================================
model = xgb.Booster()
model.load_model("xgb_model.json")

dtest = xgb.DMatrix(test)
preds = model.predict(dtest)

# =========================================================
# 제출 파일 생성
# =========================================================
submit = pd.read_csv("../sample_submission.csv")
submit["clicked"] = preds
submit.to_csv("../output/xgb_infer_dev.csv", index=False)
print("✅ 추론 완료: xgb_infer.csv 생성")
```
